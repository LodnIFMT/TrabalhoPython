```python
import pandas as pd
```

```python
planilha = pd.read_excel('listaCompletaJogos.xlsx')
display(planilha)
```

```python
jg2013 = planilha[planilha["year"] <= 2013]
display(jg2013)
```

```python
jg_ps1 = jg2013[jg2013["platform"] == "PS"]
display(jg_ps1)
```

```python
jg_ps2 = jg2013[jg2013["platform"] == "PS2"]
display(jg_ps2)
```

```python
jg_ps_ps2 = jg2013[jg2013["platform"].isin(["PS", "PS2"])]
display(jg_ps_ps2)
```

```python
lista_jogos = jg_ps_ps2.drop(columns=["id", "link",	"GameFAQs",	"bayesian avg",	"Unnamed: 11",	"Unnamed: 12",	"Unnamed: 13",	"Unnamed: 14"])
display(lista_jogos)
```

```python
lista_jogos.to_excel("Lista_Jogos.xlsx")
```
